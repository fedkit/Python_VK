## Задание 1

```python
def solve_quadratic_equation(a, b, c):
    if a == 0 and b == 0 and c == 0:
        print('x - любое число')
        return None
        
    if a == 0 and b == 0 and c != 0:
        return None
        
    if a == 0 and b != 0:
        return tuple([-c / b])
        
    disc = b**2 - 4 * a * c
    
    if disc < 0:
        return None
          
    if disc == 0:
        return tuple([-b/(2*a)])
        
    return tuple([(-b - disc**0.5) / (2 * a), (-b + disc**0.5) / (2 * a)])
```

```python
assert solve_quadratic_equation(0, 0, 0) is None
```

```python
assert solve_quadratic_equation(0, 0, -6) is None
```

```python
assert solve_quadratic_equation(0, 10, -5) == tuple([0.5])
```

```python
assert solve_quadratic_equation(0.23, 5, 101) is None
```

```python
assert solve_quadratic_equation(1, 2, 1) == tuple([-1])
```

```python
assert solve_quadratic_equation(5, 2, -7) == (-1.4, 1)
```

```python
assert solve_quadratic_equation(1, 0, -4) == (-2, 2)
```

## Задание 2

```python
def even_odd_division(l):
    even_list = tuple(filter(lambda x: x % 2 == 0, l))
    odd_list = tuple(filter(lambda x: x % 2 != 0, l))
    return even_list, odd_list
```

```python
assert even_odd_division([]) == ((), ())
```

```python
assert even_odd_division([12, 34, 6, -8]) == ((12, 34, 6, -8), ())
```

```python
assert even_odd_division([11, 887, -133, -1]) == ((), (11, 887, -133, -1))
```

```python
assert even_odd_division([9, 8, 7, 6, 5, 4]) == ((8, 6, 4), (9, 7, 5))
```

```python
assert even_odd_division(list(range(10))) == ((0, 2, 4, 6, 8), (1, 3, 5, 7, 9))
```

## Задание 3

```python
def top_k_words(text, k):
    text = text.lower()
    words = text.split()
    dct = {}
    
    for i in words:
        if i not in dct:
            dct[i] = 1
        else:
            dct[i] += 1

    answer_dop = []

    for i in dct:
        answer_dop.append((dct[i], i))

    answer_dop = sorted(answer_dop, reverse=True)
    answer = []

    for i in range(k):
        answer.append((answer_dop[i][1], answer_dop[i][0]))

    return answer
```

```python
assert top_k_words("apple apple apple", 1) == [('apple', 3)]
```

```python
assert top_k_words("a b c d a b c a", 3) == [('a', 3), ('c', 2), ('b', 2)]
```

```python
assert top_k_words("red blue green red blue red", 2) == [('red', 3), ('blue', 2)]
```

```python
test_text = """
В лесу родилась елочка,
В лесу она росла,
Зимой и летом стройная,
Зеленая была.
"""

assert top_k_words(test_text, 3) == [('лесу', 2), ('в', 2), ('стройная,', 1)]
```
